# Tensorflow Practice

```python
import numpy as np
import tensorflow as tf
```

# 1 - Practive I

$f(w) = w^2 -10w + 25$ $= (w - 5) ^ 2$

得到梯度下降最小值, 求解w ($w = 5$)

```python
# 1. 定义parameters
w = tf.Variable(0, dtype = tf.float32)
# 2. 定义损失函数
# cost = tf.add(tf.add(w**2, tf.multiply(-10, w)), 25)
cost = w**2 - 10 * w + 25
# 3. 训练函数
train = tf.compat.v1.train.GradientDescentOptimizer(0.01).minimize(cost)

# train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)


init = tf.global_variables_initializer()
session = tf.Session()
session.run(init) # 初始化全局变量
print(session.run(w)) # 评估变量, 没有训练, 所以w 是初始值
```

```python
session.run(train) # 训练一次, 梯度下降
print(session.run(w)) # 评估变量
```

```python
# 训练1000次后
for i in range(1000):
    session.run(train)

print(session.run(w)) # 评估变量
```

# 2 - Practive II

$f(w) = w^2 -10w + 25$ $= (w - 5) ^ 2$

将训练数据加入到TF中

```python
# 将该值, 带入到cost公式中
coefficients = np.array([[1.], [-10.], [25.]])

w = tf.Variable(0, dtype = tf.float32)
# 方便你去设置变量, 将数据带入
X = tf.placeholder(tf.float32, [3, 1]) # 3 * 1 matrix

# cost = w**2 - 10 * w + 25
cost = X[0][0] * w ** 2 + X[1][0] * w + X[2][0]

train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
init = tf.global_variables_initializer()
session = tf.Session()
session.run(init) # 初始化全局变量
```

```python
# 训练1000次后
for i in range(1000):
    session.run(train, feed_dict = X: coefficients)

print(session.run(w)) # 评估变量
```
